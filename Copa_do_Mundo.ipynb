**Precisamos trasformar a coluna com as datas para o formato onde é possível  filtrar os anos que ocorrem as datas. Para isso 
utilizarei o parse_dates["nome da coluna"]**

```python
dados = pd.read_csv("results.csv",  parse_dates=["date"])
dados.head(50)
```

**Agora podemos filtrar os anos, dias e meses desssas datas**

```python
# Filtrando por ano
dados['date'].dt.year
# Filtrando por mês
dados['date'].dt.month
# Filtrando por dia
dados['date'].dt.day
```

**Vamos para nosso primeiro filtro. Agora eu preciso filtrar apenas os dados correspondentes aos jogos realizados pela FIFA já que em nosso dataframe há muitos torneiros diferentes**

```python
dados_Fifa = dados[dados['tournament'] == 'FIFA World Cup']
dados_Fifa
```
