**Precisamos trasformar a coluna com as datas para o formato onde é possível  filtrar os anos que ocorrem as datas. Para isso 
utilizarei o parse_dates["nome da coluna"]**

```python
dados = pd.read_csv("results.csv",  parse_dates=["date"])
dados.head(50)
```

**Agora podemos filtrar os anos, dias e meses desssas datas**

```python
# Filtrando por ano
dados['date'].dt.year
# Filtrando por mês
dados['date'].dt.month
# Filtrando por dia
dados['date'].dt.day
```

**Vamos para nosso primeiro filtro. Agora eu preciso filtrar apenas os dados correspondentes aos jogos realizados pela FIFA já que em nosso dataframe há muitos torneios e jogos diferentes**


```python
dados_FIFA = dados[dados['tournament'] == 'FIFA World Cup']
dados_FIFA
```

**Criando uma nova coluna com as informações dos anos das partidas**

```python
dados_FIFA["year"] = dados_FIFA['date'].dt.year
dados_FIFA
```

**Agora precisamos reunir os dados dos gols de fora e de casa em um único dataframe. Para isso vamos criar dois dataframes diferentes e concatena-los**

```python
# Criando o dataframe dos gols em casa
gols_casa = dados_FIFA[["home_team", "home_score", "year"]]
gols_casa
```

```python
# Criando o dataframe dos gols fora
gols_fora = dados_FIFA[["away_team", "away_score", "year"]]
gols_fora
```

```python
# Concatenando os dataframes criados acima
scores_fifa = pd.concat([gols_casa, gols_fora])
scores_fifa
```

**Podemos ver que os dataframes foram concatenados, no entanto não é ainda a melhor maneira para visualizar dos dados. O ideial é que os dados estejam de forma vertical, sendo apenas 3 colunas. Anos, equipes e gols. Para isso vamos renomear as colunas para todas tenham o mesmo nome**

```python
gols_fora = dados_FIFA[["away_team", "away_score", "year"]].rename({'away_team':'seleção', 'away_score':'gols'},axis='columns')
gols_fora
```

```python
# Renomaandos a dataframe gols_casa
gols_casa = dados_FIFA[["home_team", "home_score", "year"]].rename({'home_team':'seleção', 'home_score':'gols'},axis='columns')
gols_casa
```

**Agora realizando o concatenação dos dataframes com os novos nome, teremos apenas 3 colunas. Seleção, gols, year**

```python
# concatenando os dataframes agora com as colunas com os nomes renomeados
gols_fifa = pd.concat([gols_fora, gols_casa]).reset_index(drop=True)
gols_fifa
```

**Para eliminar os NaN podemos utilizar o FillNa()**

```python
gols_fifa = pd.concat([gols_fora, gols_casa]).reset_index(drop=True).fillna(0)
gols_fifa
```

**Agora podemos agrupar por seleção e gols de uma maneira mais legal utilizando o método unstack(), que basicamente nos permite desemplilhar um dataframe trocando as linhas por colunas**

```python
# Utilizando o unstack para desempilhar o dataframe, ele fica parecendo uma planilha em Excel
gols_fifa.groupby(["seleção", "year"]).gols.sum().unstack().fillna(0)
```

```python
gols_fifa_saldo =gols_fifa.groupby(["seleção", "year"]).gols.sum().unstack().fillna(0).cumsum(axis=1).fillna(0).reset_index() 
gols_fifa_saldo
```

```python
bandeiras = pd.read_csv("countries-fifa-flags.csv")
bandeiras
```

**Vamos realizar o merge dos dois dataframes**

```python
gols_fifa_saldo.merge(bandeiras,left_on="seleção", right_on="country", how="left").to_csv("copa_grafico.csv", index=False)
```
