```python
import pandas as pd
import numpy as np

# Immportando dos ados
copa = pd.read_csv('WorldCups.csv')
copa
```

**Analisando os paises que mais sediaram a copa para criar um gráfico animado**

```python
# Utilizando o values_counts na coluna Country para contabilizar os paises que mais sediaram
copa["Country"].value_counts()
```

```python
copa.groupby(["Year", "Country"]).size()
```

```python
# Aqui utilizei um groupby junto com o metodo unstack() para desempilhar os dados e tornar a visualização melhor, 
# também é importante utilizar o fillna(0) para preencher os valores vazios com 0*

# Criando um dataframe com os anos e os paises sede cada ano da copa
copa.groupby(["Country", "Year"]).size().unstack().fillna(0).reset_index()
```

```python
# O mesmo dataframe anterior, porém com uma soma acumalado utilizando o cumsum(axis=1), somando por linhas
sedia_copa = copa.groupby(["Country", "Year"]).size().unstack().fillna(0).cumsum(axis=1).reset_index()
sedia_copa
```

```python
bandeiras = pd.read_csv("countries-fifa-flags.csv")
bandeiras
```

```python
# Realizando o merge com o dataframe das bandeiras com o dataframe das informações sede
sedia_copa.merge(bandeiras,left_on="Country", right_on="country", how="left").to_csv("data2.csv", index=False)
```

```python
# Criando um arquivo csv para criar o gráfico
sedia_copa.to_csv("data1.csv", index=False)
```

**2º gráfico animado: Seleções que mais chegaram na final**

Para isso tenho que criar uma coluna nova com dos dados das finais

```python
Finais_Vencedores = copa[["Winner", "Year"]].rename({"Winner":"Finalistas"},axis='columns')
Finais_Vencedores
```

```python
Finais_Vices = copa[["Runners-Up", "Year"]].rename({"Runners-Up":"Finalistas"},axis='columns')
Finais_Vices
```

```python
Finalistas = pd.concat([Finais_Vencedores, Finais_Vices]).reset_index(drop=True)
Finalistas.groupby(["Finalistas", "Year"]).size().unstack().fillna(0).cumsum(axis=1)
```

**Vamos criar um gráfico para analisar a quantidade de times qualificados ao longo dos anos de copa**

**1º Analisar os valores**

```python
teams = copa[["Year", "QualifiedTeams"]].sort_values(by="Year")
teams
```

**2 Agora vamos visualizar com um gráfico**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# teams.plot(kind="bar", x="Year")

plt.figure(figsize=(11,4))
# c = sns.color_palette("tab10")
sns.barplot(x='Year',y='QualifiedTeams', data=teams)
plt.xlabel("Anos")
plt.ylabel("Qtd seleções")
plt.title("Quantidade de seleções participantes",color='black')
plt.savefig("qts_Times.png")
plt.show()
```

```python
Matches = copa[["Year","MatchesPlayed"]]
```

```python

plt.figure(figsize=(11,4))
c = sns.color_palette("BuGn_r", 10) 
sns.barplot(x='Year',y='MatchesPlayed', data=Matches, color='c')
plt.xlabel("Anos")
plt.ylabel("Jogos Disputados")
plt.grid()
plt.title("Quantidade de jogos disputados ao longo dos anos",color='black')
plt.savefig("jogos_anos.png")
plt.show()
```
