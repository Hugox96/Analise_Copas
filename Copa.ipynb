```python
import pandas as pd
import numpy as np

copa = pd.read_csv('WorldCups.csv')
copa
```

```python
copa["Country"].value_counts().sort_values(ascending=False)
```

**Vamos criar um gráfico para analisar a quantidade de times qualificados ao longo dos anos de copa**

**1º Analisar os valores**

```python
teams = copa[["Year", "QualifiedTeams"]].sort_values(by="Year")
teams
```

**2 Agora vamos visualizar com um gráfico**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# teams.plot(kind="bar", x="Year")

plt.figure(figsize=(11,4))
c = sns.color_palette("tab10")
sns.barplot(x='Year',y='QualifiedTeams', data=teams, color='c')
plt.xlabel("Anos")
plt.ylabel("Qtd seleções")
plt.title("Quantidade de seleções participantes",color='black')
plt.show()
```
